```python
import qutip as qt
import numpy as np
from pathlib import Path
from datetime import datetime
import matplotlib as mpl
from matplotlib.ticker import MultipleLocator
import matplotlib.patheffects as pe
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from functools import lru_cache
import requests
# 通知設定
DISCORD_WEBHOOK_URL = "https://discord.com/api/webhooks/1439055708988375082/6iiXA8J_3Bn1RPWZ8XmPfmfSJp_PkGdaIzt495Ao2fbu_a09VnMJpxmrnSyfN4Wtyc7T"
def notify_discord(message: str):
    payload = {"content": message}
    try:
        r = requests.post(DISCORD_WEBHOOK_URL, json=payload, timeout=5)
        r.raise_for_status()
        print("✅ Discord 通知送信 OK")
    except Exception as e:
        print("⚠ Discord 通知に失敗:", e)

# グラフ設定
mpl.rcParams.update(mpl.rcParamsDefault)
plt.style.use('default')
mpl.rcParams.update({
    "axes.labelsize": 18,   # 軸ラベル
    "xtick.labelsize": 16,  # 目盛り数字
    "ytick.labelsize": 16,
    "lines.linewidth": 2.2, # 線の太さ
    "axes.titlesize": 16,
})

def inward_ticks(ax, major_len=7, minor_len=4, width=1.2):
    ax.minorticks_on()
    ax.tick_params(which='both', direction='in', top=True, right=True, length=major_len, width=width)
    ax.tick_params(which='minor', length=minor_len, width=width)

markers = ['o', 's', '^', 'v', 'D', '*', '+', 'x']
colors = [
    "black",
    "tab:blue",
    "tab:orange",
    "tab:green",
    "tab:purple",
    "tab:red",
    "tab:brown",
    "tab:gray",
]
OUTDIR = Path("charts")
def save_fig(fig, stem: str):
    fig.savefig(OUTDIR / f"{stem}.pdf", bbox_inches="tight", dpi=300) 
    print(f"saved: {OUTDIR / (stem + '.pdf')}")

# データ保存
DATADIR = Path("data")

# データ読み込み例
# fname = DATADIR / "square_sa_tcut.npz" 

# dat = np.load(fname)

# N_list  = dat["N"]   # shape (nN,)
# hx_list = dat["hx_list"]  # shape (nHx,)
# Tcut_list = dat["Tcut"]
# EE      = dat["EE"]       # shape (nN, nHx)
```

```python
# ========= キャッシュ付きヘルパ =========
@lru_cache(maxsize=None)
def get_ops(N):
    """Nだけで決まる演算子たち"""
    cut_pos = N // 2 - 1
    sx, sz = [0.5 * M for M in (qt.sigmax(), qt.sigmaz())]
    I2 = qt.qeye(2)

    def op_at(i, op):
        return qt.tensor([op if k == i else I2 for k in range(N)])

    Sx = [op_at(i, sx) for i in range(N)]
    Sz = [op_at(i, sz) for i in range(N)]

    def H_exchange(i):
        return Sz[i] * Sz[i+1]

    Hcut_unit   = H_exchange(cut_pos)
    Hex_rest_unit = sum(H_exchange(i) for i in range(N-1) if i != cut_pos)
    Hhx_unit    = sum(Sx[i] for i in range(N))

    left_sites = list(range(N//2))
    return Hcut_unit, Hex_rest_unit, Hhx_unit, left_sites

def simulate(N, J, hx, Tcut, tlist):
    """J, hx, Tcutだけで決まるもの"""
    Hcut_unit, Hex_rest_unit, Hhx_unit, left_sites = get_ops(N)

    H_full   = -J * (Hex_rest_unit + Hcut_unit) - hx * Hhx_unit
    # t>=Tcut のときのハミルトニアン（切断後）
    H_static = -J * Hex_rest_unit - hx * Hhx_unit

    def ramp(t, T_cut):
        if T_cut == 0.0:
            # 瞬時切断：t<=0で1, t>0で0（t=0 から見れば状態は変わらない）
            return 1.0 if t <= 0.0 else 0.0
        if t <= 0.0:
            return 1.0
        if t >= T_cut:
            return 0.0
        return 1.0 - t / T_cut

    Ht = [H_static, [-J * Hcut_unit, lambda t, args: ramp(t, Tcut)]]

    E0, psi0 = H_full.groundstate(sparse=True)

    opts = {
        "method": "bdf",
        "nsteps": 100_000_000,
        "rtol": 1e-6,
        "atol": 1e-8,
        "max_step": (Tcut / 500.0) if Tcut > 0.0 else 0.1,
        "store_states": True,
        "progress_bar": None,
    }

    res = qt.sesolve(Ht, psi0, tlist, e_ops=[], options=opts)
    
    return res
```

```python
N = 6
N_list = [N]
J = 1.0
hx = 0.067
Tcut = 3000.0

tlist = np.linspace(0.0, Tcut, 301)

res = simulate(N=N, J=J, hx=hx, Tcut=Tcut, tlist=tlist)

psi_array = np.stack([st.full().ravel() for st in res.states], axis=0)  # (nt, 2**N)
psi_array = psi_array / np.linalg.norm(psi_array, axis=1, keepdims=True)

fname = DATADIR / "tfim_en_jp_n6_hx0.067.npz"
dat = np.load(fname, allow_pickle=True)

Jp_list   = dat["Jp_list"]
psin_list = dat["psin_list"]

eig_idx = [0, 1, 2, 3, 4] 
n_eig = len(eig_idx)
nt = len(tlist)

def ramp_Jp(t: float, J: float, Tcut: float) -> float:
    """simulate 内の J'(t) と同じ関数に合わせること！"""
    if t <= 0.0:
        return J
    elif t >= Tcut:
        return 0.0
    else:
        return J * (1.0 - t / Tcut)

Jp_of_t = np.array([ramp_Jp(t, J, Tcut) for t in tlist])   
idx_Jp  = np.array([np.argmin(np.abs(Jp_list - Jp)) for Jp in Jp_of_t]) 

static_vecs = psin_list[idx_Jp][:, eig_idx, :]  

static_vecs = static_vecs / np.linalg.norm(static_vecs, axis=2, keepdims=True)

fid_tn = np.abs(np.einsum("tnd,td->tn", np.conjugate(static_vecs), psi_array))

fid = fid_tn.T 

fig, ax = plt.subplots(figsize=(6.9, 4.6))
inward_ticks(ax)

for i, idx in enumerate(eig_idx):
    ax.plot(
        tlist,
        fid[i],
        marker=markers[i],
        color=colors[i],
        markersize=6.0,
        markevery=10
    )

# ax.set_xlim()

ax.set_xlim(0.0, 3000.0)
ax.set_xlabel(r"$t$", fontsize=24)
ax.set_ylim(0.0, 1.05)
ax.set_ylabel(r"$|\langle \phi_n (J'(t)) | \psi(t) \rangle|$", fontsize=18)

plt.show()

npz_path = DATADIR / f"tfim_fid_vs_t_n{N}.npz"
np.savez(
    npz_path,
    N=N,
    tlist=tlist,
    eig_idx=np.array(eig_idx),
    fid=fid,
)

notify_discord(message=f"tifm sim 4-1 done")
save_fig(fig=fig, stem=f"tfim_fid_vs_t_n{N}")
```
